```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

from matplotlib import pyplot as plt
from datetime import datetime as dt
```

```python
df = pd.read_csv('./full_data.csv')
```

```python
df = df[df['location'] == 'Russia']

date_func = np.vectorize(lambda x: dt.strptime(x, "%Y-%m-%d"))
df = df[date_func(df['date']) >= dt.strptime('2020-03-03', "%Y-%m-%d")]
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(np.arange(len(df.iloc[:,2])), df.iloc[:,2], label='new cases')
plt.legend()
plt.subplot(1,2,2)
plt.plot(np.arange(len(df.iloc[:,2])), df.iloc[:,4], label='total cases')
plt.legend()
plt.show()
```

$\text{cases} = \exp (a\ \text{day} + b) \ \  \ \ \ln \text{cases} = a\ \text{day} + b$

Начнём с максимизации правдоподобия

```python
X = np.arange(50)[:,None]
y = np.log(np.maximum(np.array(df['total_cases'][:50]),1))
regressor = LinearRegression()
regressor.fit(X, y)

X_pred = np.arange(70)[:,None]
y_pred = regressor.predict(X_pred)

plt.figure()
plt.plot(np.arange(len(df.iloc[:,2])), df.iloc[:,4], label='total cases')
plt.plot(X_pred, np.exp(y_pred), label='predicted total cases')
plt.legend()
plt.show()
```

Отсюда оценим дисперсию $\sigma^2$ шума:

```python
sigm2 = (y_pred[:len(y)] - y).var()
print('sigma^2: ', sigm2)
```

В качестве априорного распределения параметров $a, b$ возьмём
\begin{equation*}
p(a,b) \propto \mathcal{N}(a|0,0.5) \mathcal{N}(b|0,5), \ \ a \geq 0
\end{equation*}
Тогда
\begin{equation*}
p(a,b|\text{data}) \propto \mathcal{N}\left( a,b| \mu, \Sigma \right), \ \ a \geq 0,
\end{equation*}
где вектор $\mu$ и матрицу $\Sigma$ легко получить численно:

```python
Sigma_0 = np.array([[0.5,0],[0,5]])

X_ext = np.concatenate([X, np.ones_like(X)], axis=1)

Sigma = np.linalg.inv(np.linalg.inv(Sigma_0) +\
                                 (1/sigm2)*(X_ext[:,None,:]*X_ext[:,:,None]).sum(axis=0))
mu = (1/sigm2)*np.matmul(Sigma, (y[:,None]*X_ext).sum(axis=0))
```

```python
def post_sample(mu, Sigma):
    while True:
        a, b = np.random.multivariate_normal(mu, Sigma)
        if a >=0:
            break
    return a, b
```

```python
plt.figure()
for i in range(5):
    a, b = post_sample(mu, Sigma)
    plt.plot(X, np.exp(a*X + b), label = f"sample#{i}")
plt.plot(X, np.exp(y), label = "true")
plt.legend()
plt.show()
```

```python
def empirical_dist(day, mu, Sigma, n_samples=1000):
    predictions = []
    for _ in range(n_samples):
        a, b = post_sample(mu, Sigma)
        predictions.append(np.exp(a*day + b))
    return predictions
```

```python
first_of_may = (dt.strptime('2020-05-01', "%Y-%m-%d") - dt.strptime('2020-03-03', "%Y-%m-%d")).days
first_of_june = (dt.strptime('2020-06-01', "%Y-%m-%d") - dt.strptime('2020-03-03', "%Y-%m-%d")).days
first_of_sept = (dt.strptime('2020-09-01', "%Y-%m-%d") - dt.strptime('2020-03-03', "%Y-%m-%d")).days

plt.figure(figsize=(18, 4))
plt.subplot(1,3,1)
plt.title('first of May dist')
plt.hist(empirical_dist(first_of_may, mu, Sigma))
plt.subplot(1,3,2)
plt.title('first of June dist')
plt.hist(empirical_dist(first_of_june, mu, Sigma))
plt.subplot(1,3,3)
plt.title('first of Sept dist')
plt.hist(empirical_dist(first_of_sept, mu, Sigma))
plt.show()
```

Теперь рассмотрим модель
\begin{equation*}
\text{cases}(\text{day}) = \frac{1}{\sqrt{2\pi}} \int\limits_{-\infty}^{\text{day}} e^{-\left(a t^2 + b t +c\right) } dt
\end{equation*}
Можно заметить, что
\begin{equation*}
\text{new_cases}(\text{day}) = \text{cases}(\text{day}) - \text{cases}(\text{day} - 1) = \
    \int\limits_{\text{day} - 1}^{\text{day}} e^{-\left(a t^2 + b t +c\right) } dt \approx \
    \frac{1}{\sqrt{2\pi}} e^{-\left( a\ (\text{day} - 1/2)^2 + b\ (\text{day} - 1/2) + c \right)}
\end{equation*}
Тогда можно обучать линейную модель
\begin{equation*}
-\sqrt{2\pi} \log \text{new_cases}(\text{day}) = a\ (\text{day} - 1/2)^2 + b\ (\text{day} - 1/2) + c
\end{equation*}

Начнём снова с максимального правдоподобия:

```python
X = np.stack([(np.arange(49) + .5)**2, (np.arange(49) + .5), np.ones(49)], axis=1)
y = - np.sqrt(2*np.pi)*np.log(np.maximum(np.array(df['new_cases'])[1:50], 1))

regressor = LinearRegression(fit_intercept=False)
regressor.fit(X, y)

X_pred = np.stack([(np.arange(69) + .5)**2, (np.arange(69) + .5), np.ones(69)], axis=1)
y_pred = regressor.predict(X_pred)
```

```python
def new_cases_from_y(y):
    return np.exp(-(1/np.sqrt(2*np.pi))*y)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title('new_cases')
plt.plot(np.arange(49) + 1, new_cases_from_y(y), label = 'true new cases')
plt.plot(np.arange(69) + 1, new_cases_from_y(y_pred), label = 'predicted new cases')
plt.legend()
plt.subplot(1,2,2)
plt.title('total_cases')
plt.plot(np.arange(49) + 1, df['total_cases'][1:50], label = 'true total cases')
plt.plot(np.arange(69) + 1, np.cumsum(new_cases_from_y(y_pred)), label = 'predicted total cases')
plt.legend()
plt.show()
```

Выглядит гораздо лучше, чем экспонента!

```python
sigm2 = (y_pred[:len(y)] - y).var()
print('sigma^2: ', sigm2)
```

В качестве априорного распределения возьмём
\begin{equation*}
    p(a,b,c) \propto \mathcal{N}\left(a,b,c|0, I)\right), \ \ a > 10^{-3}
\end{equation*}
Апостериорное распределение снова будет нормальным с параметрами $\mu, \Sigma$

```python
Sigma_0 = np.diag(np.ones(3))

Sigma = np.linalg.inv(np.linalg.inv(Sigma_0) +\
                                 (1/sigm2)*(X[:,None,:]*X[:,:,None]).sum(axis=0))
mu = (1/sigm2)*np.matmul(Sigma, (y[:,None]*X).sum(axis=0))
```

```python
def post_sample(mu, Sigma):
    while True:
        a, b, c = np.random.multivariate_normal(mu, Sigma)
        if a > 0.001:
            break
    return a, b, c
```

```python
plt.figure()
days = np.arange(1, 200)
for i in range(5):
    a, b, c = post_sample(mu, Sigma)
    plt.plot(days, np.cumsum(np.exp(-(a*(days - 0.5)**2 + b*(days - 0.5) + c)/np.sqrt(2*np.pi))),
                 label = f"sample#{i}")
plt.legend()
plt.show()
```

```python
def empirical_dist(day, mu, Sigma, n_samples=1000):
    predictions = []
    days = np.arange(1, day)
    for _ in range(n_samples):
        a, b, c = post_sample(mu, Sigma)
        predictions.append(np.sum(np.exp(-(a*(days - 0.5)**2 + b*(days - 0.5) + c)/np.sqrt(2*np.pi))))
    return predictions
```

```python
first_of_may = (dt.strptime('2020-05-01', "%Y-%m-%d") - dt.strptime('2020-03-03', "%Y-%m-%d")).days
first_of_june = (dt.strptime('2020-06-01', "%Y-%m-%d") - dt.strptime('2020-03-03', "%Y-%m-%d")).days
first_of_sept = (dt.strptime('2020-09-01', "%Y-%m-%d") - dt.strptime('2020-03-03', "%Y-%m-%d")).days

plt.figure(figsize=(18, 4))
plt.subplot(1,3,1)
plt.title('first of May dist')
plt.hist(empirical_dist(first_of_may, mu, Sigma))
plt.subplot(1,3,2)
plt.title('first of June dist')
plt.hist(empirical_dist(first_of_june, mu, Sigma))
plt.subplot(1,3,3)
plt.title('first of Sept dist')
plt.hist(empirical_dist(first_of_sept, mu, Sigma))
plt.show()
```

То же для разных стран

```python
def get_post_parameters(new_cases_data):
    X = np.stack([(np.arange(len(new_cases_data)-1) + .5)**2,
                    (np.arange(len(new_cases_data) - 1) + .5),
                    np.ones(len(new_cases_data) - 1)], axis=1)
    y = - np.sqrt(2*np.pi)*np.log(np.maximum(new_cases_data[1:], 1))
    
    regressor = LinearRegression(fit_intercept=False)
    regressor.fit(X, y)
    
    y_pred = regressor.predict(X)
    
    sigm2 = (y - y_pred).var()

    Sigma_0 = np.diag(np.ones(3))

    Sigma = np.linalg.inv(np.linalg.inv(Sigma_0) +\
                                 (1/sigm2)*(X[:,None,:]*X[:,:,None]).sum(axis=0))
    mu = (1/sigm2)*np.matmul(Sigma, (y[:,None]*X).sum(axis=0))
    
    return mu, Sigma
```

Ширина и высота колокола, стоящего под интегралом, даются комбинациями $\frac{1}{\sqrt{2a}},\ \frac{b^2}{4a^2} - c$

```python
dist_params = {}
df = pd.read_csv('./full_data.csv')

dist_params['Russia'] = get_post_parameters(np.array(df[df['location'] == 'Belarus']['new_cases'][63:]))
dist_params['Belarus'] = get_post_parameters(np.array(df[df['location'] == 'Belarus']['new_cases'][63:]))
dist_params['Germany'] = get_post_parameters(np.array(df[df['location'] == 'Germany']['new_cases'][57:]))
dist_params['United States'] = \
    get_post_parameters(np.array(df[df['location'] == 'United States']['new_cases'][58:]))
dist_params['Italy'] = get_post_parameters(np.array(df[df['location'] == 'Italy']['new_cases'][53:]))
dist_params['United Kingdom'] = \
    get_post_parameters(np.array(df[df['location'] == 'United Kingdom']['new_cases'][63:]))
dist_params['Spain'] = get_post_parameters(np.array(df[df['location'] == 'Spain']['new_cases'][57:]))
dist_params['India'] = get_post_parameters(np.array(df[df['location'] == 'India']['new_cases'][75:]))
dist_params['Sweden'] = get_post_parameters(np.array(df[df['location'] == 'Sweden']['new_cases'][65:]))
dist_params['Switzerland'] = \
    get_post_parameters(np.array(df[df['location'] == 'Switzerland']['new_cases'][60:]))
dist_params['Ukraine'] = get_post_parameters(np.array(df[df['location'] == 'Ukraine']['new_cases'][5:]))
dist_params['Russia'] = get_post_parameters(np.array(df[df['location'] == 'Russia']['new_cases'][73:]))
```

```python
dots = []
names = []
for key, val in dist_params.items():
    width = 1/np.sqrt(2*val[0][0])
    height = val[0][1]**2/(4*val[0][0]**2) - val[0][2]
    dots.append([width, height])
    names.append(key)
dots = np.array(dots)
```

```python
plt.scatter(dots[:,0], dots[:,1])
for i, name in enumerate(names):
    plt.annotate(name, (dots[i,0], dots[i,1]))
plt.show()
```
